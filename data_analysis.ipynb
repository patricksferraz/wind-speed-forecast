```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def show_all(dataframe, index):
    nans = np.where(np.isnan(dataframe.values))
    print(
        "\nColumn {}\nMax {}\nMédia {}\nMediana {}\nMínimo {}\nNan {}".format(
            index,
            dataframe.max(),
            dataframe.mean(),
            dataframe.median(),
            dataframe.min(),
            np.where(np.isnan(dataframe.values))
            # dataframe.mode(),
        )
    )
    return nans
```

```python
dataframe = pd.read_csv(
    "confidential/train150.txt", delimiter="\t", header=None
)
dataset = dataframe.values
```

```python
# plot the speeds
dataframe_t = pd.read_csv("confidential/train150.txt", delimiter="\t", header=None)
dataset_t = dataframe_t.values
dataframe_p = pd.read_csv("confidential/prev150.txt", delimiter="\t", header=None)
dataset_p = dataframe_p.values
YY = np.concatenate([dataset_t[:, 4], dataset_p[:,4]])
y_size = range(len(YY))

print(len(dataset_p))

plt.figure(figsize=(20,10))
plt.style.use("ggplot")
plt.plot(y_size, YY, label="Série temporal")
plt.plot(y_size, [YY.mean()] * len(YY), linestyle="--", label="Velocidade média")
plt.title("Velocidade média do vento em Mucuri (novembro a dezembro de 2015)")
plt.xlabel("Tempo (h)")
plt.ylabel("Velocidade do vento (m/s)")
plt.legend()
plt.show()
```
